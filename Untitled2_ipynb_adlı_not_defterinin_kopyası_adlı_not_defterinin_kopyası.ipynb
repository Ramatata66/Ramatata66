```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
import numpy as np

# Sahibinden'den alınan motosiklet verisi (manuel)
motosikletler = {
    'Marka': ['Yamaha', 'Honda', 'Suzuki', 'Ducati', 'Kawasaki'],
    'Model': ['MT-07', 'CBR500R', 'GSX250R', 'Monster', 'Ninja 400'],
    'Yıl': [2020, 2019, 2018, 2022, 2021],
    'Fiyat': [150000, 135000, 95000, 250000, 160000]
}

# DataFrame'e dönüştürme
motosiklet_df = pd.DataFrame(motosikletler)

# X ve y belirleme
X = motosiklet_df[['Yıl']]
y = motosiklet_df['Fiyat']

# Regresyon modeli oluşturma ve eğitme
model = LinearRegression()
model.fit(X, y)

# Regresyon katsayılarını yazdırma
print(f"Eğim (Slope): {model.coef_[0]}")
print(f"Y-ekseni kesişimi (Intercept): {model.intercept_}")

# Tahmin yapma
y_pred = model.predict(X)

# Scatter plot ve regresyon doğrusu
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Yıl', y='Fiyat', data=motosiklet_df, s=100, label='Gerçek Fiyatlar')
plt.plot(motosiklet_df['Yıl'], y_pred, color='red', label='Regresyon Doğrusu')

# Grafik detayları
plt.title('Yıl ile Fiyat Arasındaki Regresyon Analizi')
plt.xlabel('Yıl')
plt.ylabel('Fiyat')
plt.legend()
plt.grid(True)
plt.show()

# R² skoru hesaplama
r2_score = model.score(X, y)
print(f"R² Skoru: {r2_score:.2f}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm  # Standart hata hesaplama için

# Sahibinden'den alınan motosiklet verisi (manuel)
motosikletler = {
    'Marka': ['Yamaha', 'Honda', 'Suzuki', 'Ducati', 'Kawasaki'],
    'Model': ['MT-07', 'CBR500R', 'GSX250R', 'Monster', 'Ninja 400'],
    'Yıl': [2020, 2019, 2018, 2022, 2021],
    'Fiyat': [150000, 135000, 95000, 250000, 160000]
}

# DataFrame'e dönüştürme
motosiklet_df = pd.DataFrame(motosikletler)

# X ve y değişkenleri
X = motosiklet_df[['Yıl']]
y = motosiklet_df['Fiyat']

# 1. Sklearn ile Regresyon Modeli
model = LinearRegression()
model.fit(X, y)

# Katsayıları ve Y-eksenini kesişimi
print(f"Eğim (Slope): {model.coef_[0]}")
print(f"Y-ekseni kesişimi (Intercept): {model.intercept_}")

# 2. Statsmodels ile Standart Hata ve Detaylı Analiz
X_sm = sm.add_constant(X)  # Sabit ekleme (Intercept)
sm_model = sm.OLS(y, X_sm).fit()  # OLS Modeli kurma

# Model özetini yazdırma
print(sm_model.summary())

# 3. Gerçek Fiyatlar ve Regresyon Doğrusu ile Görselleştirme
y_pred = model.predict(X)  # Modelin tahmin ettiği fiyatlar

plt.figure(figsize=(10, 6))
sns.scatterplot(x='Yıl', y='Fiyat', data=motosiklet_df, s=100, label='Gerçek Fiyatlar')
plt.plot(motosiklet_df['Yıl'], y_pred, color='red', label='Regresyon Doğrusu')

# Grafik Detayları
plt.title('Yıl ile Fiyat Arasındaki Regresyon Analizi')
plt.xlabel('Yıl')
plt.ylabel('Fiyat')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Gerekli kütüphaneleri yükleme
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import matplotlib.pyplot as plt
import seaborn as sns

# Iris veri setini yükleme
from sklearn.datasets import load_iris
iris = load_iris()
X = iris.data
y = iris.target

# Veriyi eğitim ve test olarak %70 - %30 bölelim
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Modellerin tanımlanması
models = {
    "Lojistik Regresyon": LogisticRegression(max_iter=200),
    "Karar Ağacı": DecisionTreeClassifier(),
    "K-En Yakın Komşu (KNN)": KNeighborsClassifier(),
    "Destek Vektör Makineleri (SVM)": SVC()
}

# Her model için performans değerlendirme metriklerini tutacağımız bir sözlük
model_performance = {}

# Modellerin eğitilmesi ve test edilmesi
for model_name, model in models.items():
    # Model eğitimi
    model.fit(X_train, y_train)
    # Tahminler
    y_pred = model.predict(X_test)

    # Performans değerlendirmesi
    accuracy = accuracy_score(y_test, y_pred)
    cm = confusion_matrix(y_test, y_pred)
    report = classification_report(y_test, y_pred, target_names=iris.target_names)

    # Performansı sözlüğe ekleme
    model_performance[model_name] = {
        "Doğruluk": accuracy,
        "Karışıklık Matrisi": cm,
        "Sınıflandırma Raporu": report
    }

    # Sonuçları ekrana yazdırma
    print(f"\n{model_name} Modeli:")
    print(f"Doğruluk: {accuracy:.2f}")
    print("Karışıklık Matrisi:\n", cm)
    print("Sınıflandırma Raporu:\n", report)

# Doğruluk skorlarını bir grafik ile karşılaştıralım
accuracy_scores = [model_performance[model]["Doğruluk"] for model in model_performance.keys()]
model_names = list(model_performance.keys())

plt.figure(figsize=(10, 6))
sns.barplot(x=model_names, y=accuracy_scores, palette="viridis")
plt.title("Modellerin Doğruluk Skorları Karşılaştırması")
plt.xlabel("Model")
plt.ylabel("Doğruluk Skoru")
plt.ylim(0, 1)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import fetch_openml, load_iris, load_digits

# 1. Veri Setlerini Yükleme
# MNIST veri seti
mnist = fetch_openml('mnist_784', version=1, as_frame=False)
mnist_images = mnist.data / 255.0  # Normalize et
mnist_sample = mnist_images[:100]  # İlk 100 görüntüyü al

# Iris veri seti
iris = load_iris()
iris_data = iris.data  # 150 örnek, 4 özellik
iris_sample = iris_data[:100]  # İlk 100 örneği al

# Digits veri seti
digits = load_digits()
digits_data = digits.data / 16.0  # Normalizasyon (piksel değerleri 0-1 arasında olmalı)
digits_sample = digits_data[:100]  # İlk 100 örneği al

# 2. PCA ile sıkıştırma ve yeniden yapılandırma
n_components = 2  # Sıkıştırma boyutunu 2 yapalım, daha kolay görselleştirilebilir

# PCA'yı her veri seti için uygulayacağız
def pca_compression_and_reconstruction(data, n_components):
    pca = PCA(n_components=n_components)
    compressed = pca.fit_transform(data)  # Sıkıştırma
    reconstructed = pca.inverse_transform(compressed)  # Yeniden yapılandırma
    return compressed, reconstructed

# Her veri seti için PCA'yı uyguluyoruz
mnist_compressed, mnist_reconstructed = pca_compression_and_reconstruction(mnist_sample, n_components)
iris_compressed, iris_reconstructed = pca_compression_and_reconstruction(iris_sample, n_components)
digits_compressed, digits_reconstructed = pca_compression_and_reconstruction(digits_sample, n_components)

# 3. Görselleştirme
# MNIST Görselleştirme
plt.figure(figsize=(12, 8))

# MNIST Görseli
plt.subplot(3, 2, 1)
plt.imshow(mnist_sample[0].reshape(28, 28), cmap="gray")
plt.title("MNIST Orijinal Görüntü")
plt.axis("off")

plt.subplot(3, 2, 2)
mnist_reconstructed_image = mnist_reconstructed[0].reshape(28, 28)
plt.imshow(mnist_reconstructed_image, cmap="gray")
plt.title(f"MNIST Yeniden Yapılandırılmış ({n_components} Bileşen)")
plt.axis("off")

# Iris Görselleştirme
plt.subplot(3, 2, 3)
plt.scatter(iris_sample[:, 0], iris_sample[:, 1], c='blue', label='Iris Orijinal')
plt.title("Iris Orijinal")
plt.xlabel('Özellik 1')
plt.ylabel('Özellik 2')

plt.subplot(3, 2, 4)
plt.scatter(iris_compressed[:, 0], iris_compressed[:, 1], c='red', label='Iris PCA Sıkıştırma')
plt.title(f"Iris PCA Sıkıştırma ({n_components} Bileşen)")
plt.xlabel('PCA Bileşeni 1')
plt.ylabel('PCA Bileşeni 2')

# Digits Görselleştirme
plt.subplot(3, 2, 5)
plt.imshow(digits_sample[0].reshape(8, 8), cmap="gray")
plt.title("Digits Orijinal Görüntü")
plt.axis("off")

plt.subplot(3, 2, 6)
digits_reconstructed_image = digits_reconstructed[0].reshape(8, 8)
plt.imshow(digits_reconstructed_image, cmap="gray")
plt.title(f"Digits Yeniden Yapılandırılmış ({n_components} Bileşen)")
plt.axis("off")

# Görselleri Göster
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import fetch_openml, load_iris, load_digits
from sklearn.metrics import mean_squared_error

# 1. Veri Setlerini Yükleme
# MNIST veri seti
mnist = fetch_openml('mnist_784', version=1, as_frame=False)
mnist_images = mnist.data / 255.0  # Normalize et
mnist_sample = mnist_images[:100]  # İlk 100 görüntüyü al

# Iris veri seti
iris = load_iris()
iris_data = iris.data  # 150 örnek, 4 özellik
iris_sample = iris_data[:100]  # İlk 100 örneği al

# Digits veri seti
digits = load_digits()
digits_data = digits.data / 16.0  # Normalizasyon (piksel değerleri 0-1 arasında olmalı)
digits_sample = digits_data[:100]  # İlk 100 örneği al

# 2. PCA ile sıkıştırma ve yeniden yapılandırma
n_components = 2  # Sıkıştırma boyutunu 2 yapalım, daha kolay görselleştirilebilir

# PCA'yı her veri seti için uygulayacağız
def pca_compression_and_reconstruction(data, n_components):
    pca = PCA(n_components=n_components)
    compressed = pca.fit_transform(data)  # Sıkıştırma
    reconstructed = pca.inverse_transform(compressed)  # Yeniden yapılandırma
    return compressed, reconstructed

# Her veri seti için PCA'yı uyguluyoruz
mnist_compressed, mnist_reconstructed = pca_compression_and_reconstruction(mnist_sample, n_components)
iris_compressed, iris_reconstructed = pca_compression_and_reconstruction(iris_sample, n_components)
digits_compressed, digits_reconstructed = pca_compression_and_reconstruction(digits_sample, n_components)

# Rekonstrüksiyon Kayıplarını hesaplamak
def calculate_reconstruction_loss(original, reconstructed):
    return mean_squared_error(original, reconstructed)

# Her veri seti için rekonstrüksiyon kaybını hesaplıyoruz
mnist_loss = calculate_reconstruction_loss(mnist_sample, mnist_reconstructed)
iris_loss = calculate_reconstruction_loss(iris_sample, iris_reconstructed)
digits_loss = calculate_reconstruction_loss(digits_sample, digits_reconstructed)

# 3. Görselleştirme
# MNIST Görselleştirme
plt.figure(figsize=(12, 8))

# MNIST Görseli
plt.subplot(3, 2, 1)
plt.imshow(mnist_sample[0].reshape(28, 28), cmap="gray")
plt.title("MNIST Orijinal Görüntü")
plt.axis("off")

plt.subplot(3, 2, 2)
mnist_reconstructed_image = mnist_reconstructed[0].reshape(28, 28)
plt.imshow(mnist_reconstructed_image, cmap="gray")
plt.title(f"MNIST Yeniden Yapılandırılmış ({n_components} Bileşen)\nRekonstrüksiyon Kayıp: {mnist_loss:.4f}")
plt.axis("off")

# Iris Görselleştirme
plt.subplot(3, 2, 3)
plt.scatter(iris_sample[:, 0], iris_sample[:, 1], c='blue', label='Iris Orijinal')
plt.title("Iris Orijinal")
plt.xlabel('Özellik 1')
plt.ylabel('Özellik 2')

plt.subplot(3, 2, 4)
plt.scatter(iris_compressed[:, 0], iris_compressed[:, 1], c='red', label='Iris PCA Sıkıştırma')
plt.title(f"Iris PCA Sıkıştırma ({n_components} Bileşen)\nRekonstrüksiyon Kayıp: {iris_loss:.4f}")
plt.xlabel('PCA Bileşeni 1')
plt.ylabel('PCA Bileşeni 2')

# Digits Görselleştirme
plt.subplot(3, 2, 5)
plt.imshow(digits_sample[0].reshape(8, 8), cmap="gray")
plt.title("Digits Orijinal Görüntü")
plt.axis("off")

plt.subplot(3, 2, 6)
digits_reconstructed_image = digits_reconstructed[0].reshape(8, 8)
plt.imshow(digits_reconstructed_image, cmap="gray")
plt.title(f"Digits Yeniden Yapılandırılmış ({n_components} Bileşen)\nRekonstrüksiyon Kayıp: {digits_loss:.4f}")
plt.axis("off")

# Görselleri Göster
plt.tight_layout()
plt.show()

# Rekonstrüksiyon Kayıplarını Yazdır
print(f"MNIST Rekonstrüksiyon Kayıp (MSE): {mnist_loss:.4f}")
print(f"Iris Rekonstrüksiyon Kayıp (MSE): {iris_loss:.4f}")
print(f"Digits Rekonstrüksiyon Kayıp (MSE): {digits_loss:.4f}")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# 4. PCA Sonrası k-NN ile Sınıflandırma

# MNIST için
mnist_labels = mnist.target[:100]  # İlk 100 görüntünün etiketlerini al
mnist_train, mnist_test, mnist_labels_train, mnist_labels_test = train_test_split(
    mnist_compressed, mnist_labels, test_size=0.2, random_state=42
)

knn_mnist = KNeighborsClassifier(n_neighbors=3)
knn_mnist.fit(mnist_train, mnist_labels_train)
mnist_predictions = knn_mnist.predict(mnist_test)
mnist_accuracy = accuracy_score(mnist_labels_test, mnist_predictions)

# Iris için
iris_labels = iris.target[:100]  # İlk 100 örneğin etiketlerini al
iris_train, iris_test, iris_labels_train, iris_labels_test = train_test_split(
    iris_compressed, iris_labels, test_size=0.2, random_state=42
)

knn_iris = KNeighborsClassifier(n_neighbors=3)
knn_iris.fit(iris_train, iris_labels_train)
iris_predictions = knn_iris.predict(iris_test)
iris_accuracy = accuracy_score(iris_labels_test, iris_predictions)

# Digits için
digits_labels = digits.target[:100]  # İlk 100 örneğin etiketlerini al
digits_train, digits_test, digits_labels_train, digits_labels_test = train_test_split(
    digits_compressed, digits_labels, test_size=0.2, random_state=42
)

knn_digits = KNeighborsClassifier(n_neighbors=3)
knn_digits.fit(digits_train, digits_labels_train)
digits_predictions = knn_digits.predict(digits_test)
digits_accuracy = accuracy_score(digits_labels_test, digits_predictions)

# 5. Sonuçları Yazdırma
print(f"MNIST k-NN Sınıflandırma Doğruluğu: {mnist_accuracy:.4f}")
print(f"Iris k-NN Sınıflandırma Doğruluğu: {iris_accuracy:.4f}")
print(f"Digits k-NN Sınıflandırma Doğruluğu: {digits_accuracy:.4f}")
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# MNIST için
kmeans_mnist = KMeans(n_clusters=3, random_state=42)
mnist_clusters = kmeans_mnist.fit_predict(mnist_train)
mnist_silhouette = silhouette_score(mnist_train, mnist_clusters)

# Iris için
kmeans_iris = KMeans(n_clusters=3, random_state=42)
iris_clusters = kmeans_iris.fit_predict(iris_train)
iris_silhouette = silhouette_score(iris_train, iris_clusters)

# Digits için
kmeans_digits = KMeans(n_clusters=3, random_state=42)
digits_clusters = kmeans_digits.fit_predict(digits_train)
digits_silhouette = silhouette_score(digits_train, digits_clusters)

# Sonuçları Yazdırma
print(f"MNIST Silhouette Skoru: {mnist_silhouette:.4f}")
print(f"Iris Silhouette Skoru: {iris_silhouette:.4f}")
print(f"Digits Silhouette Skoru: {digits_silhouette:.4f}")
```

```python
import numpy as np
import tensorflow as tf
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import precision_score, recall_score, accuracy_score, f1_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, Flatten, Dense
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.applications import VGG16
from tensorflow.keras.datasets import mnist, imdb
import pandas as pd

# Sonuçları toplamak için liste
results = []

### MNIST Veri Seti (Görüntü)
(X_train_img, y_train_img), (X_test_img, y_test_img) = mnist.load_data()
X_train_img, X_test_img = X_train_img[:500] / 255.0, X_test_img[:100] / 255.0  # Küçük veri seti
y_train_img, y_test_img = y_train_img[:500], y_test_img[:100]
X_train_img_flat = X_train_img.reshape(-1, 28 * 28)
X_test_img_flat = X_test_img.reshape(-1, 28 * 28)

# Logistic Regression
lr_model = LogisticRegression(max_iter=200)
lr_model.fit(X_train_img_flat, y_train_img)
y_pred_lr = lr_model.predict(X_test_img_flat)
results.append([
    "Logistic Regression (MNIST)",
    precision_score(y_test_img, y_pred_lr, average='weighted'),
    recall_score(y_test_img, y_pred_lr, average='weighted'),
    accuracy_score(y_test_img, y_pred_lr),
    f1_score(y_test_img, y_pred_lr, average='weighted'),
])

# SVM
svm_model = SVC()
svm_model.fit(X_train_img_flat, y_train_img)
y_pred_svm = svm_model.predict(X_test_img_flat)
results.append([
    "SVM (MNIST)",
    precision_score(y_test_img, y_pred_svm, average='weighted'),
    recall_score(y_test_img, y_pred_svm, average='weighted'),
    accuracy_score(y_test_img, y_pred_svm),
    f1_score(y_test_img, y_pred_svm, average='weighted'),
])

# Daha Küçük ve Hızlı CNN Modeli
cnn = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
    Flatten(),
    Dense(128, activation='relu'),
    Dense(10, activation='softmax')
])
cnn.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
cnn.fit(X_train_img[..., None], y_train_img, epochs=3, batch_size=32, verbose=0)
y_pred_cnn = np.argmax(cnn.predict(X_test_img[..., None]), axis=1)
results.append([
    "CNN (MNIST)",
    precision_score(y_test_img, y_pred_cnn, average='weighted'),
    recall_score(y_test_img, y_pred_cnn, average='weighted'),
    accuracy_score(y_test_img, y_pred_cnn),
    f1_score(y_test_img, y_pred_cnn, average='weighted'),
])

# VGG16 (Transfer Learning) - Önceden Eğitilmiş Model Kullanımı
vgg16 = VGG16(weights='imagenet', include_top=False, input_shape=(32, 32, 3))  # RGB'ye uygun hale getirildi

# Gri tonlama görüntüleri 3 kanala dönüştürme
X_train_img_resized = tf.image.resize_with_pad(X_train_img[..., None], 32, 32).numpy()
X_test_img_resized = tf.image.resize_with_pad(X_test_img[..., None], 32, 32).numpy()

# VGG16 modelinin özellik çıkarıcı kısmını kullanıyoruz
X_train_img_rgb = np.repeat(X_train_img_resized, 3, axis=-1)  # 3 kanala kopyalayarak dönüştürme
X_test_img_rgb = np.repeat(X_test_img_resized, 3, axis=-1)    # 3 kanala kopyalayarak dönüştürme

# VGG16'ın sadece özellik çıkarma kısmını kullanıyoruz
vgg16_features_train = vgg16.predict(X_train_img_rgb)
vgg16_features_test = vgg16.predict(X_test_img_rgb)

# Logistic Regression ile sınıflandırma
lr_model_vgg16 = LogisticRegression(max_iter=200)
lr_model_vgg16.fit(vgg16_features_train.reshape(500, -1), y_train_img)
y_pred_vgg16 = lr_model_vgg16.predict(vgg16_features_test.reshape(100, -1))
results.append([
    "VGG16 + Logistic Regression (MNIST)",
    precision_score(y_test_img, y_pred_vgg16, average='weighted'),
    recall_score(y_test_img, y_pred_vgg16, average='weighted'),
    accuracy_score(y_test_img, y_pred_vgg16),
    f1_score(y_test_img, y_pred_vgg16, average='weighted'),
])

### IMDB Veri Seti (Metin)
(X_train_text, y_train_text), (X_test_text, y_test_text) = imdb.load_data(num_words=5000)
X_train_text = pad_sequences(X_train_text[:200], maxlen=100)  # Küçük veri seti
X_test_text = pad_sequences(X_test_text[:40], maxlen=100)
y_train_text = y_train_text[:200]
y_test_text = y_test_text[:40]

# Logistic Regression for Text
lr_model_text = LogisticRegression(max_iter=200)
lr_model_text.fit(X_train_text, y_train_text)
y_pred_text_lr = lr_model_text.predict(X_test_text)
results.append([
    "Logistic Regression (IMDB)",
    precision_score(y_test_text, y_pred_text_lr, average='weighted'),
    recall_score(y_test_text, y_pred_text_lr, average='weighted'),
    accuracy_score(y_test_text, y_pred_text_lr),
    f1_score(y_test_text, y_pred_text_lr, average='weighted'),
])

# Sonuç Tablosu
df_results = pd.DataFrame(results, columns=["Model", "Precision", "Recall", "Accuracy", "F1-Score"])
print(df_results)
```
